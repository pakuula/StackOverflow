```python
import numpy as np
import scipy.optimize as spo
import matplotlib.pyplot as plt
from scipy.stats import norm
import random as rnd
import numba
```

```python
_factor_norm = np.sqrt(2*np.pi)
def priceNormFn(x):
    return _factor_norm*norm.pdf(x)

def priceFn(x):
    v = x/2
    v2=v*v
    return 1.0 - v2/np.sqrt(1+v2*v2)
```

```python
# Штраф для функции притяжения - потенциальная яма, обращающаяся в 0 на бесконечности
# На большом расстоянии уже пофиг, что там притягивает.
def attract(power, distance):
    return -power*priceFn(distance/(power+1.0))
```

```python
# Штраф для отталкивания - постепенно убывает, обращается в ноль на бесконечности
def repell(power, distance):
    return power*priceFn(distance/(power+1.0))
```

```python
fig, axs = plt.subplots(1,2)
fig.set_size_inches(11,3)
x = np.linspace(0,10,100)
axs[0].plot(x, attract(1, x))
axs[1].plot(x, repell(1, x))
```

```python
class Point:
    def __init__(self, coords, attracting, repelling):
        self.coords = np.array(coords)
        self.attracting_power = attracting
        self.repelling_power = repelling
    def potential(self, v):
        dv = self.coords - np.atleast_1d(v)
        distance = np.linalg.norm(dv)
        attracting_penalty = attract(self.attracting_power, distance) if self.attracting_power > 0 else 0.0
        repelling_penalty = repell(self.repelling_power, distance) if self.repelling_power > 0 else 0.0
        return attracting_penalty + repelling_penalty
    def __repr__(self):
        return f"Point({tuple(self.coords)}, {self.attracting_power}, {self.repelling_power})"
```

```python
supermarket = Point((0.0, 0.0), attracting=10, repelling=0.5)
X = np.linspace(-10,10,100)
plt.plot(X, [supermarket.potential((_x,0)) for _x in X])
```

```python
class Potential:
    def __init__(self, points):
        self.points = points
    
    @numba.jit(forceobj=True)
    def potential(self, v):
        resPot = 0.0
        for p in self.points:
            potential = p.potential(v) 
            resPot += potential
        return resPot
    def __call__(self, v):
        return self.potential(v)
```

# Гораздо более быстрый вариант

Время счёта потенциала можно значительно, в 500 раз, улучшить, если все вычисления выполнять только над структурами numpy и откомпилировав функции в машинный код JIT компилятором numba.

Вычисления функции оценивания в 10 тысячах точек для поля, создаваемого 100  центрами. С использованием классов `Potential` и `Points` заняло 11 секунд, а JIT-откомпилированная функции - 24 миллисекунды.

```python
def mk_v(x,y):
    return x+y*1.0j

@numba.njit(inline="always")
def distance(v1, v2) -> float :
    return abs(v1-v2)

@numba.njit(inline="always")
def _attract(power, v0, v1):
    d = distance(v0,v1)/2
    d = d/(abs(power) +1)
    d2 = d*d
    return -power*(1.0-d2/np.sqrt(1+d2*d2))
        
@numba.njit(inline="always")
def point_potential(point, v):
    v0, power = point
    return _attract(power, v0, v)

@numba.njit(inline='always')
def group_potential(points, v):
    pot = [_attract(power,v0, v) for v0, power in points]
    return np.real(np.sum(np.array(pot)))

@numba.njit
def potential(points, vv):
    return np.array([group_potential(points, v) for v in vv])
```
